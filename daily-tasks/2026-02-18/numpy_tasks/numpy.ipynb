
2. Given two 2D arrays of different shapes, use broadcasting to compute their outer sum (no loops).

```python
arr1 = np.array([[1,2,3],
                 [4,5,6]])
# 2,3
arr2 = np.array([[1],[2]])
# 2,1

arr1 = arr1[:,:,np.newaxis, np.newaxis]
arr2 = arr2[np.newaxis, np.newaxis, :,:]

result = arr1 + arr2
print(result)
print(result.shape)
```

3. Create an array of 100 random numbers from a normal distribution and clip all values outside 2 standard deviations.

```python
normal_arr = np.random.randn(100)

std_arr = np.std(normal_arr)

result = np.clip(normal_arr, -2*std_arr, 2*std_arr)

print(result)
print("Min:", result.min(), "Max:", result.max())
```


7. Stack three 1D arrays of length 5 vertically and horizontally, then describe the shape of each result.

```python
arr1 = np.random.rand(5)
arr2 = np.random.rand(5)
arr3 = np.random.rand(5)

hstack_arr = np.hstack((arr1,arr2,arr3))
vstack_arr = np.vstack((arr1,arr2,arr3))
print(hstack_arr)
print(hstack_arr.shape)

print(vstack_arr)
print(vstack_arr.shape)
```


8. Compute the cumulative sum of an array row-wise and column-wise using np.cumsum().

```python
arr = np.array([[1,2],[3,4]])
print( np.cumsum(arr, axis=1) ) # row-wise
print( np.cumsum(arr, axis=0) ) # column-wise
```


9. Given a 2D array, find all rows where the row sum exceeds 100.

```python
arr = np.array([[10,50,50],
          [1,2,3],
          [40,40,30]])

# print(np.where(np.sum(arr, axis=1) > 100))
print(arr[np.sum(arr, axis=1) > 100])
```
